```python
import gym
```

```python
env = gym.make('CartPole-v0')
```

# 환경 초기화 
env.reset()

```python
random_episode = 0
rewardsTotal = 0
```

```python
while random_episode < 10:
    env.render()
    action = env.action_space.sample()
    # observe = state, 
    observe, reward, done,_ = env.step(action)
    print(observe,reward,done)
    rewardsTotal += reward
    if done:
        random_episode += 1
        print("에피소드 보상", rewardsTotal)
        rewardsTotal = 0
        env.reset()
```
